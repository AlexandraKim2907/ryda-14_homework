```python
# рассчитаем моду
def moda_calculate ():
    surgery_counts = {}
    for x in req_data.surgery.round():
        if x not in surgery_counts:
            surgery_counts[x] = 1
        else:
            surgery_counts[x] += 1
            
    max_x = 0
    mode_surgery = None
    for k, v in surgery_counts.items():
        if max_x < v:
            max_x = v
            mode_surgery = k
    print('Значение моды для surgery:', mode_surgery, 'количество встречаемости:', max_x)
    
    age_counts = {}
    for x in req_data.age.round():
        if x not in age_counts:
            age_counts[x] = 1
        else:
            age_counts[x] += 1
            
    max_x = 0
    mode_age = None
    for k, v in age_counts.items():
        if max_x < v:
            max_x = v
            mode_age = k
    print ('Значение моды для age:', mode_age, 'количество встречаемости:', max_x)
    
    temperature_counts = {}
    for x in req_data['rectal temperature'].round():
        if x not in temperature_counts:
            temperature_counts[x] = 1
        else:
            temperature_counts[x] += 1
            
    max_x = 0
    mode_temperature = None
    for k, v in temperature_counts.items():
        if max_x < v:
            max_x = v
            mode_temperature = k
    print('Значение моды для rectal temperature:', mode_temperature, 'количество встречаемости:', max_x)
    
    pulse_counts = {}
    for x in req_data.pulse.round():
        if x not in pulse_counts:
            pulse_counts[x] = 1
        else:
            pulse_counts[x] += 1
            
    max_x = 0
    mode_pulse = None
    for k, v in pulse_counts.items():
        if max_x < v:
            max_x = v
            mode_pulse = k
    print('Значение моды для pulse:', mode_pulse, 'количество встречаемости:', max_x)
    
    abdominal_distension_counts = {}
    for x in req_data['abdominal distension'].round():
        if x not in abdominal_distension_counts:
            abdominal_distension_counts[x] = 1
        else:
            abdominal_distension_counts[x] += 1
            
    max_x = 0
    mode_abdominal_distension = None
    for k, v in abdominal_distension_counts.items():
        if max_x < v:
            max_x = v
            mode_abdominal_distension = k
    print('Значение моды для abdominal distension:', mode_abdominal_distension, 'количество встречаемости:', max_x)
    
    abdomcentesis_total_protein_counts = {}
    for x in req_data['abdomcentesis total protein'].round():
        if x not in abdomcentesis_total_protein_counts:
            abdomcentesis_total_protein_counts[x] = 1
        else:
            abdomcentesis_total_protein_counts[x] += 1
    
    max_x = 0
    mode_abdomcentesis_total_protein = None
    for k, v in abdomcentesis_total_protein_counts.items():
        if max_x < v:
            max_x = v
            mode_abdomcentesis_total_protein = k
    print('Значение моды для abdomcentesis total protein:', mode_abdomcentesis_total_protein, 'количество встречаемости:', max_x)
    
    outcome_counts = {}
    for x in req_data.outcome.round():
        if x not in outcome_counts:
            outcome_counts[x] = 1
        else:
            outcome_counts[x] += 1
    
    max_x = 0
    mode_outcome = None
    for k, v in outcome_counts.items():
        if max_x < v:
            max_x = v
            mode_outcome = k
    print('Значение моды для outcome:', mode_outcome, 'количество встречаемости:', max_x)
    
    type_of_lesion_counts = {}
    for x in req_data['type of lesion(25)'].round():
        if x not in type_of_lesion_counts:
            type_of_lesion_counts[x] = 1
        else:
            type_of_lesion_counts[x] += 1
    
    max_x = 0
    mode_type_of_lesion = None
    for k, v in type_of_lesion_counts.items():
        if max_x < v:
            max_x = v
            mode_type_of_lesion = k
    print('Значение моды для type of lesion(25):', mode_type_of_lesion, 'количество встречаемости:', max_x)
    
moda_calculate()
```
